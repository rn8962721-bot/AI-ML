```python
def is_even(num):
    """
    This function return if a give number is odd or even
    input - any valid integer
    output - odd/even
    create on  - 24-02-26
    """
    if type(num) == int:
        if num % 2 == 0:
            print("Even")
        else:
            print("Odd")
    else:
        return "enter the number correctly"
```

```python
for i in range(1,11):
    x = is_even(i)
    print(x)
```
