```python
# Set Operation
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 & s2

s1 - s2
s2 - s1

s1 ^ s2

1 not in s1

for i in s1: 
    print(i)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 | s2

s1.update(s2)

print(s1)
print(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2)

s1.difference_update(s2)

print(s1)
print(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```
