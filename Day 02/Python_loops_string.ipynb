```python
s = input("Enter the String")
flag = True
for i in range(0,len(s)//2):
    if s[i] != s[len(s) -i -1]:
        flag = False
        print("Not a Palindrome")
        break
    if flag:
        print("Palindrome")
```

```python
s = input("Enter a Sting ")

L = []
temp = ''
for i in s:
    if i != ' ':
        temp = temp + i
    else:
        L.append(temp)
        temp = ''
    
L.append(temp)
print(L)
```

```python
s = input("Enter the Sting")

L = []
for i in s.split():
    L.append(i[0].upper() + i[1:].lower())

print(''.join(L))
```

```python
number = int(input("Enter the Number"))

digits = '0123456789'
result = ''
while number != 0:
    result = digits[number % 10] + result
    number = number//10

print(number)
print(type(number))
```
